```python
import numpy as np

# Create a list
data_list = [10, 20, 30, 40, 50]

# Convert list to NumPy array
arr = np.array(data_list)

# Display array, shape, and data type
print("Array:", arr)
print("Shape:", arr.shape)
print("Data Type:", arr.dtype)
```

```python
import numpy as np

# Create 3x4 matrix of zeros
zeros_matrix = np.zeros((3, 4))

# Create 2x5 matrix of ones
ones_matrix = np.ones((2, 5))

print("Zeros Matrix:\n", zeros_matrix)
print("Shape of Zeros Matrix:", zeros_matrix.shape)

print("\nOnes Matrix:\n", ones_matrix)
print("Shape of Ones Matrix:", ones_matrix.shape)
```

```python
import numpy as np

# Create array from 1 to 12
arr = np.arange(1, 13)

# Reshape into 3x4 matrix
reshaped_arr = arr.reshape(3, 4)

print("Original Array:", arr)
print("Original Shape:", arr.shape)

print("\nReshaped Matrix:\n", reshaped_arr)
print("Reshaped Shape:", reshaped_arr.shape)
```

```python
import numpy as np

# Generate 10 random integers between 1 and 100
random_arr = np.random.randint(1, 100, 10)

print("Random Array:", random_arr)
print("Minimum:", random_arr.min())
print("Maximum:", random_arr.max())
print("Mean:", random_arr.mean())
print("Sum:", random_arr.sum())
```

```python
import pandas as pd

# Create a dictionary
data = {
    'Student': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Vikram'],
    'Marks': [85, 92, 78, 88, 76]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame and Data Types
print(df)
print("\nData Types:\n", df.dtypes)
```

```python
import pandas as pd

# Create a dummy CSV file for demonstration
data = {
    'ID': range(1, 16),
    'Name': ['Student' + str(i) for i in range(1, 16)],
    'Score': [70, 80, 90, 65, 75, 85, 95, 60, 72, 82, 92, 68, 78, 88, 98]
}
pd.DataFrame(data).to_csv('students.csv', index=False)

# Read the CSV file
df = pd.read_csv('students.csv')

print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())
```
